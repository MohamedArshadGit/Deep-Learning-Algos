```python
import tensorflow,keras
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
voc_size=10000
```

```python
#for personal check
it=0
for i in sent:
    print(i)
    it+=1
    print(it)
```

```python
onehotrep=[one_hot(input_text=word,n=voc_size)for word in sent]
print(onehotrep)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
```

```python
sentences_length=10
pre_embedded=pad_sequences(sequences=onehotrep,padding='pre',maxlen=sentences_length)
#if we didnt give sentences length it will automatically assign according to maximum lenghted sentence
```

```python
print(pre_embedded)
```

```python
#we need to initialise dimension for embedding layer(user defined only)
dimension=10 
```

```python
model=Sequential()
```

```python
model.add(Embedding(input_dim=voc_size,output_dim=dimension,input_length=sentences_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(pre_embedded)
```
